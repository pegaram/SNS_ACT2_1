```python
# LIBRERÍAS EMPLEADAS EN EL EJEMPLO
# ==============================================================================

# Importamos las librerías necesarias
import numpy as np  # Numpy es una librería para operaciones numéricas y manejo de arrays
import pandas as pd  # Pandas es una librería para manipulación de datos, especialmente en estructuras como DataFrames

# Generación de datos
# ==============================================================================

# Generamos 100 elementos de ingresos uniformemente distribuidos en [15000, 25000]
# np.random.rand(100) genera 100 números aleatorios entre 0 y 1, distribuidos de forma uniforme
# Al multiplicar por 10000, los números generados estarán en el rango [0, 10000]
# Sumamos 15000 a estos valores para trasladarlos al rango [15000, 25000]
datos = 15000 + np.random.rand(100) * 10000  # Generamos un array de 100 elementos en el rango [15000, 25000]

# Se hace una copia de los datos generados y se ordenan de menor a mayor
# datos_ord es una copia de los datos originales
# Luego, con el método .sort(), ordenamos los valores en orden ascendente.
datos_ord = datos  # Copia de los datos originales
datos_ord.sort()  # Ordenamos los datos de menor a mayor

# Finalmente, mostramos el array de los datos generados
# El resultado 'datos' contiene los 100 elementos generados y 'datos_ord' es la misma secuencia ordenada
datos  # Mostramos los datos generados sin ordenar
```

#**Ejercicio 1.b**: Aplicar el método de Probabilidad global, para detectar los outliers utilizado en el ejemplo 2_3_Outliers

```python
# CRITERIO 1: PROBABILIDAD GLOBAL - Explicado en el capítulo 5 - Preprocesamiento (Semana 2), en la pág 10
# Asumiendo que las variables tiene una distribución normal.
# Probabilidad de la muestra de estar dentro de las bandas
p_g=0.95
# probabilidad global
alfa_g=(1-p_g)/2
# probabilidad  para un solo dato
alfa= 1-(1-alfa_g)**(1/len(datos)) # Se realiza este ajuste para ser más precisos.
#alfa = alfa_g # Si no realiza la corrección de la línea anterior entonces hubiese detectado más valores que son outliers cuando realmente no lo son. Descomentar esta línea para ver el resultado, que es el mismo que el obtenidoi con el método 2 de este código.
                # con esta corrección lo que se pretende es obtener bandas lo sificientemente anchas.
```

```python
# Impresión de resultados
alfa=round(alfa,5) # Redondeamos el valor de alfa a 5 decimales
print(f" Alfa ={alfa}") # Imprimimos el valor de alfa redondeado
```

```python
# Utilizaremos el Criterio 1: PROBABILIDAD GLOBAL
# Calcular los límites utilizando la media
xL=  round(np.mean(datos)-alfa*  np.std(datos),4) # Calculamos el límite inferior de la banda
xU=  round(np.mean(datos)+alfa*  np.std(datos),4) # Calculamos el límite superior de la banda

# Calculamos la diferencia entre los límites superior e inferior de la banda
diferencia= xU-xL

# Imprimimos los resultados
print(f" La media es {np.mean(datos)}")
print(f" Banda= [ {xL},{xU}]")
print(f" El ancho de la Banda es {diferencia}")
```

```python
# Inicializamos contadores para los datos dentro de los límites y fuera de los límites
contador_dentro = 0
contador_fuera = 0

# Recorremos los datos
for i in range(len(datos)):
    if xL <= datos[i] <= xU:
        contador_dentro += 1  # Incrementar contador de datos dentro de los límites
    else:
        contador_fuera += 1  # Incrementar contador de outliers (fuera de los límites)

# Imprimir resultados
print(f"El número de elementos dentro de los límites [{xL}, {xU}] es: {contador_dentro}")
print(f"El número de outliers fuera de los límites [{xL}, {xU}] es: {contador_fuera}")
```

El 100% de los datos utilizando la media son Outliers
